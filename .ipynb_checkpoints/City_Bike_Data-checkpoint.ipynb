## Citi Bike Rides History Data

```python
# Create list of links to Citi Bike Rides History Data
links = []
links_JC = []
path = 'https://s3.amazonaws.com/tripdata/'
year_start = 2013 
year_end = 2020 
# loop at year
for y in range(year_start, year_end, 1):
    i = 1
    j = 13
    
    if y == 2013:
        i = 6
        
    if y == 2019:
        j = 10
      
    csv_name = ''
    if y > 2016:
        csv_name = '.csv'
    
# loop at month    
    for m in range(i, j, 1):
        s = '0'
        if m > 9:
          s = ''

        links.append(f'{path}{y}{s}{m}-citibike-tripdata{csv_name}.zip')  
        
        if (y == 2015 and m >= 9) or y > 2015:
            links.append(f'{path}JC-{y}{s}{m}-citibike-tripdata{csv_name}.zip')
            links_JC.append(f'{path}JC-{y}{s}{m}-citibike-tripdata{csv_name}.zip')

    if y == 2013:
        links.append(path + '201307-201402-citibike-tripdata.zip')  
    
    if y == 2016:
        links.append(path + '201611-citibike-tripdata 23-17-16-417.zip')
        
    if y == 2017:    
        links.append(path + '201702-citibike-tripdata.csv 23-17-18-428.zip')
        links.append(path + '201704-citibike-tripdata.csv 23-17-24-762.zip')  
        
    if y == 2018:
        links.append(path + '201810-citibike-tripdata.csv 23-18-13-331.zip')
        
```

```python
# total links list 
print(links) 
```

```python
# total quantity of files
len(links)
```

```python
# quantity of JC files
len(links_JC)
```

```python
# list of links to JC files
for l in links_JC:
    print(l)
```

```python
# Read data from the JC files to the list of dataframes
rides = []
for link in links_JC:
    try:
        df = pd.read_csv(link, low_memory=False)
        rides.append(df)            
    except: continue  
        
```

```python
# quantity of data frames
print(len(rides))
```

```python
# check the column names
for i,r in enumerate(rides):
    print(f'df{i}:')
    print(rides[i].columns)
```

```python
# rename colum names of first three data frames
for i in range(0,3,1):
    rides[i].columns = ['tripduration', 'starttime', 'stoptime', 'start station id',
       'start station name', 'start station latitude',
       'start station longitude', 'end station id', 'end station name',
       'end station latitude', 'end station longitude', 'bikeid', 'usertype',
       'birth year', 'gender']
    print(f'df{i}:')
    print(rides[i].columns)
```

```python
# Create Total Data Frame fo JC rides history since 2017 until 2019
rides_df = pd.concat(rides) 
```

```python
# detirmine missed data
def missing_data(df):
    total = df.isnull().sum().sort_values(ascending=False)
    percent = (df.isnull().sum()/df.isnull().count()*100).sort_values(ascending=False)
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    return missing_data
missing_data(rides_df)
```

```python
# change columns types
rides_df['starttime'] = pd.to_datetime(rides_df['starttime'])
rides_df['stoptime'] = pd.to_datetime(rides_df['stoptime'])
rides_df['start station name'] = rides_df['start station name'].astype('category')
rides_df['end station name'] = rides_df['end station name'].astype('category')
rides_df['usertype'] = rides_df['usertype'].astype('category')
rides_df['gender'] = rides_df['gender'].astype('category')
rides_df['birth year'] = round(rides_df['birth year'],0)
#round(rides_df,2).head(10)
```
